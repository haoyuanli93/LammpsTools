```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 14})
```

```python
thermoInfo = np.loadtxt("./myThermo.*.txt.")

########################################################################################
fig, axes = plt.subplots(ncols=3, nrows=1)

fig.set_figheight(6)
fig.set_figwidth(18)

window_size = 10
start_idx = 0#1500

axes[0].plot(thermoInfo[start_idx:,0], thermoInfo[start_idx:,2], 'b')
axes[0].set_xlabel("step")
axes[0].set_ylabel("temp (K)")
axes[0].set_ylim([600, 630])
#axes[0].set_xlim([0, 200])
axes[0].set_title("Temperature: {:.2f} K".format(np.mean(thermoInfo[-window_size:-1, 2])))

axes[1].plot(thermoInfo[start_idx:,0], thermoInfo[start_idx:,1], 'b')
axes[1].set_xlabel("step")
axes[1].set_ylabel("Press (bar)")
axes[1].set_ylim([-0, 200])
axes[1].set_title("Pressure: {:.2f} bar".format(np.mean(thermoInfo[-window_size:-1, 1])))

axes[2].plot(thermoInfo[start_idx:,0], thermoInfo[start_idx:,3], 'b')
axes[2].set_xlabel("step")
axes[2].set_ylabel("Density g/cm^3")
axes[2].set_title("Density: {:.2f} g/cm^3".format(np.mean(thermoInfo[-window_size:-1, 3])))

plt.tight_layout()
plt.show()
```

```python
tem_fft = np.fft.fftshift(np.fft.fft(thermoInfo[start_idx:,2]))
pres_fft = np.fft.fftshift(np.fft.fft(thermoInfo[start_idx:,1]))
```

```python
plt.plot(np.square(np.abs(tem_fft)))
plt.yscale("log")
plt.show()
```

```python
plt.plot(np.square(np.abs(pres_fft)))
plt.yscale("log")
plt.show()
```
